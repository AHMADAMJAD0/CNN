```python
import tensorflow
from tensorflow import keras
from keras.layers import Conv2D , Flatten , Dense , MaxPooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(x_train , y_train) , (x_test , y_test) = mnist.load_data()
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3) , padding = ('valid') , activation = 'relu' , input_shape = (28 , 28 , 1))) # padding valid mean do the image same as dont use padding
model.add(MaxPooling2D(pool_size = (2,2) , strides = 2 , padding = "valid"))
model.add(Conv2D(32 , kernel_size = (3,3) , padding = "valid" , activation = 'relu')) # 32 mean filters which learn pattrens and 3,3 mean 3 by 3 hota hai wo
model.add(MaxPooling2D(pool_size = (2,2) , strides = 2 , padding = "valid"))
# model.add(Conv2D(32 , kernel_size = (3,3) , padding = 'valid' , activation = 'relu' ))
model.add(Flatten())
model.add(Dense(128 , activation = 'relu'))
model.add(Dense(10 , activation = 'softmax'))
```

```python
model.summary()
```
